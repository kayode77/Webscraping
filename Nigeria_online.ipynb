# Nigeria From Wikipedia, the free encyclopedia

Nigeria, officially the Federal Republic of Nigeria, is a country in West Africa. It is situated between the Sahel to the north and the Gulf of Guinea to the south in the Atlantic Ocean. It covers an area of 923,769 square kilometres (356,669 sq mi), and with a population of over 230 million, it is the most populous country in Africa, and the world's sixth-most populous country. Nigeria borders Niger in the north, Chad in the northeast, Cameroon in the east, and Benin in the west. Nigeria is a federal republic comprising 36 states and the Federal Capital Territory, where the capital, Abuja, is located. The largest city in Nigeria is Lagos, one of the largest metropolitan areas in the world and the largest in Africa.

```python
from bs4 import BeautifulSoup as bs
import requests
r=requests.get('https://en.wikipedia.org/wiki/Nigeria')
# convert to beautiful soup object
soup=bs(r.content)
# print out the HTML
contents = soup.prettify()
print(contents)
```

```python
info_row=soup.find_all("tr")
Nig_info={}
for index,row in enumerate(info_row):
    if index == 0:
        Nig_info['title']=row.find('th').get_text(' ', strip=True).replace('\xa0',' ')
print(Nig_info)
```

```python
def Extract_Table(url):
    r=requests.get(url)
    # convert to beautiful soup object
    soup=bs(r.content)
    Table = soup.find_all('th',class_='infobox-label')
    Tablehead=[c.get_text(' ', strip=True).replace('\xa0',' ').replace('â¢ ',' ') for c in Table]
    Tabled = soup.find_all('td',class_='infobox-data')
    Tabledata=[c.get_text(' ', strip=True).replace('\xa0',' ').replace('\ufeff',' ').replace(' /',',') for c in Tabled]
    a=zip(Tablehead,Tabledata)
    for index,x in enumerate(a):
        print(index + 1 ,x)
```

```python
# Calling the function to extract the side Table
url= 'https://en.wikipedia.org/wiki/Nigeria'
Extract_Table(url)
```

# Another Method and Presenting it in Json Format

```python
import json

data_dict1 = {}

def Extract_Table1(url):
    r=requests.get(url)
    # convert to beautiful soup object
    soup=bs(r.content)
    # Find all rows in the table
    table_rows = soup.find_all('tr')
    # Iterate over each row and extract data
    for row in table_rows:
        label = row.find('th', class_='infobox-label')
        data = row.find('td', class_='infobox-data')
        if label and data:
            data_dict1[label.get_text(' ', strip=True).replace('\xa0',' ').replace('â¢ ',' ')] = data.get_text(' ', strip=True).replace('\xa0',' ').replace('\ufeff',' ').replace(' /',',')
    # Convert Python to JSON  
    json_object1 = json.dumps(data_dict1, indent = 4) 
    # Print JSON object
    print(json_object1) 
            
```

```python
# Calling the function to extract the side Table
url= 'https://en.wikipedia.org/wiki/Nigeria'
Extract_Table1(url)
```

# Method three(for such a website with in consistency in tr,th and td)
## All the information will be scrapped from the site containing table head(tr) and Table data (td) by using the Table row(tr)

```python
import json

data_dict2 = {}

def Extract_Table2(url):
    r=requests.get(url)
    # convert to beautiful soup object
    soup=bs(r.content)
    # Find all rows in the table
    rows = soup.find_all('tr')
    # Initialize an empty dictionary to store data
    # Loop through each row
    for row in rows:
        # Find the cells in the head(th) and the data(td) by using the row
        cells = row.find_all(['th', 'td'])
        # Extract the label and data
        if len(cells) == 2:
            label = cells[0].get_text(' ', strip=True).replace('\xa0',' ').replace('â¢ ',' ')
            data = cells[1].get_text(' ', strip=True).replace('\xa0',' ').replace('\ufeff',' ').replace(' /',',')
            # Store label and data in the dictionary
            data_dict2[label] = data
    # Convert Python to JSON  
    json_object1 = json.dumps(data_dict2, indent = 4) 
    # Print JSON object
    print(json_object1) 

```

```python
# Calling the function to extract the side Table
url= 'https://en.wikipedia.org/wiki/Nigeria'
Extract_Table2(url)
```
